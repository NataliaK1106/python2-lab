## Ćwiczenie 1: Series — tworzenie i operacje 

```python
import pandas as pd
import numpy as np
# Series z listy — domyślny indeks (0, 1, 2, ...)
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

```python
#1a Tworzenie Series pensje ze słownika

# Klucze: 'Anna', 'Jan', 'Ewa', 'Marek', 'Kasia'
# Wartości: 5500, 7200, 4800, 9100, 6300
pensje = pd.Series({
    'Anna' : 5500, 
    'Jan' : 7200, 
    'Ewa' : 4800, 
    'Marek' : 9100, 
    'Kasia' : 6300
    })
print(pensje)
```

```python
#1b Pytania o pensję

# 1. Wyświetl imię osoby zarabiającej najwięcej (użyj idxmax())
najwiecej = pensje.idxmax()
print(najwiecej)

# 2. Oblicz średnią pensję
srednia = pensje.mean()
print(srednia)

# 3. Wyświetl imiona osób zarabiających powyżej średniej (filtrowanie boolean)

powyzej_sredniej = pensje[pensje > srednia]
print(powyzej_sredniej)

# 4. Policz ile osób zarabia poniżej 6000 (użyj len() lub .sum())
ponizej_6000 = pensje < 6000

liczba_osob = ponizej_6000.sum()
print(liczba_osob)
```

```python
#1c Pensja po zmianach

# 1. Oblicz pensje po podwyżce 10% dla wszystkich (pensje * 1.10)
pensja_podwyżka = pensje * 1.10

# 2. Oblicz pensje po premii 500 zł dla wszystkich (pensje + 500)
pensja_premia = pensje + 500

# 3. Oblicz nową pensję = (pensja * 1.10) + 500
nowa_pensja = (pensje * 1.10) + 500

# 4. Wyświetl ranking nowych pensji od najwyższej (użyj sort_values)
pensje_posortowane = nowa_pensja.sort_values(ascending=False)
print(pensje_posortowane)
```

## Ćwiczenie 3: Eksploracja datasetu — samodzielna analiza

```python
#3a Pierwszy rzut oka

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)

# 1. Wyświetl pierwsze 10 wierszy (head)
print(penguins.head(10))

# 2. Wyświetl kształt tabeli (shape)
print(penguins.shape)

# 3. Uruchom info() — ile brakujących wartości widzisz?
print(penguins.info())
```

```python
#3b Statystyki opisowe

# 1. Uruchom describe() — statystyki numeryczne
print(penguins.describe())

# 2. Uruchom describe(include='all') — co dodają kolumny tekstowe?
print(penguins.describe(include='all'))

# 3. Mediana masy ciała (body_mass_g)
mediana_masa = penguins['body_mass_g'].median()
print("Mediana body_mass_g:", mediana_masa)
```

```python
#3c Rozkłady wartości

# 1. Ile pingwinów każdego gatunku? (species)
print(penguins['species'].value_counts())

# 2. Ile pingwinów na każdej wyspie? (island)
print(penguins['island'].value_counts())

# 3. Rozkład płci (sex) — czy jest zbalansowany?
print(penguins['sex'].value_counts())

# 4. Jaki procent to gatunek Adelie? (normalize=True)
print(penguins['species'].value_counts(normalize=True) * 100)
```

```python
#3d Brakujące dane

# 1. Policz brakujące wartości w każdej kolumnie
print(penguins.isna().sum())

# 2. Jaki procent danych brakuje w kolumnie 'sex'?
braki_sex = penguins['sex'].isna().sum()
procent_brak = braki_sex / len(penguins) * 100
print("Procent braków w kolumnie 'sex':", procent_brak, "%")
```

## Ćwiczenie 4: Pytania biznesowe + commit

```python
#4a Pytania biznesowe — dataset tips

import pandas as pd

# Wczytanie datasetu
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
#1. Oblicz średni rachunek (total_bill)
sredni_rachunek = tips['total_bill'].mean()
print("Średni rachunek:", round(sredni_rachunek, 2))
```

```python
#2. Oblicz średni napiwek w procentach (tip_pct)
sredni_tip_pct = tips['tip_pct'].mean()
print("Średni napiwek (%):", round(sredni_tip_pct, 1))
```

```python
#3. Wyświetl który dzień ma najwięcej rachunków (value_counts)
najwiecej_dzien = tips['day'].value_counts().idxmax()
print("Dzień z największą liczbą rachunków:", najwiecej_dzien)
```

```python
#4. Policz ile rachunków jest od palaczy (value_counts na 'smoker')
rachunki_palacze = tips['smoker'].value_counts()
print("Rachunki od palaczy i niepalących:\n", rachunki_palacze)
```

```python
#5. Wyświetl describe() tylko dla kolumny 'tip'
print(tips['tip'].describe())

# Mediana napiwku
mediana_tip = tips['tip'].median()
print("Mediana napiwku:", mediana_tip)
```

```python
#6. Jaki procent rachunków pochodzi z obiadu (Dinner)?
procent_dinner = tips['time'].value_counts(normalize=True)['Dinner'] * 100
print("Procent rachunków Dinner:", round(procent_dinner, 1), "%")
```

```python
#7. Ile unikalnych rozmiarów grup jest w danych? (nunique na 'size')
unikalne_rozmiary = tips['size'].nunique()
print("Liczba unikalnych rozmiarów grup:", unikalne_rozmiary)
```
