# Ćwiczenie 3:

```python
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

## Zadanie 1: Podstawowe statystyki cenowe
Średnia cena, najdroższy i najtańszy produkt

```python
# Zadanie 1: Podstawowe statystyki cenowe

# Średnia cena
srednia_cena = ceny.mean()
# Najdroższy produkt
najdrozszy_ind = ceny.argmax()
najdrozszy = nazwy[najdrozszy_ind]
# Najtańszy produkt
najtanszy_ind = ceny.argmin()
najtanszy = nazwy[najtanszy_ind]

print(f"Średnia cena: {srednia_cena} zł")
print(f"Najdroższy produkt: {najdrozszy} ({ceny[najdrozszy_ind]} zł)")
print(f"Najtańszy produkt: {najtanszy} ({ceny[najtanszy_ind]} zł)")
```

## Zadanie 2: Wartość magazynu
Oblicz wartość każdego produktu i łączną wartość magazynu

```python
# Zadanie 2: wartość magazynu
 
# Wartość każdego produktu
wartosc_produktow = ceny * ilosci_magazyn

# Tworzymy słownik: nazwa -> wartość
slownik_wartosci = dict(zip(nazwy, wartosc_produktow))

# Łączna wartość magazynu
laczna_wartosc = wartosc_produktow.sum()

# Wypisanie wartości każdego produktu
print("Wartość magazynowa poszczególnych produktów:")
for produkt, wartosc in slownik_wartosci.items():
    print(f"{produkt}: {wartosc} zł")

# Wypisanie łącznej wartości
print(f"\nŁączna wartość magazynu: {laczna_wartosc} zł")
```

## Zadanie 3: Analiza cenowa
Stwórz array z cenami po rabacie 15% (promocja świąteczna) oraz z uwzględnieniem VAT 23%.  
Policz, ile produktów kosztuje powyżej 1000 zł po rabacie.

```python
# Zadanie 3: Analiza cenowa
rabat = ceny * 0.85
vat = ceny * 1.23

# Ile produktów kosztuje powyżej 1000 zł po rabacie?
liczba_pow_1000 = (rabat > 1000).sum()

print("Ceny po rabacie 15%:", rabat)
print("Ceny po uwzględnieniu VAT 23%:", vat)
print(f"Liczba produktów po rabacie > 1000 zł: {liczba_pow_1000}")
```

## Zadanie 4: Filtrowanie
Znajdź produkty:  
- z oceną >= 4.5 — wyświetl nazwy i ceny  
- poniżej 200 zł — wyświetl nazwy i ilości w magazynie

```python
# Zadanie 4: Filtrowanie
nazwy_array = np.array(nazwy)

# Produkty z oceną >= 4.5
mask_4_5 = oceny_klientow >= 4.5
print("Produkty z oceną >= 4.5:")
for n, c in zip(nazwy_array[mask_4_5], ceny[mask_4_5]):
    print(f"{n}: {c} zł")

# Produkty poniżej 200 zł
mask_200 = ceny < 200
print("\nProdukty < 200 zł i ilość w magazynie:")
for n, il in zip(nazwy_array[mask_200], ilosci_magazyn[mask_200]):
    print(f"{n}: {il} szt.")
```

## Zadanie 5: Analiza zapasów
Wyświetl produkty z niskim stanem magazynowym (<30 szt.),  
oblicz ich łączną wartość i procent całego magazynu.

```python
# Zadanie 5: Analiza zapasów
mask_niski = ilosci_magazyn < 30
wartosci_niskie = (ceny * ilosci_magazyn)[mask_niski]
suma_niskie = wartosci_niskie.sum()
procent_magazynu = suma_niskie / (ceny * ilosci_magazyn).sum() * 100

print("Produkty z niskim stanem magazynowym (<30 szt.):")
for n in nazwy_array[mask_niski]:
    print(n)
print(f"Łączna wartość tych produktów: {suma_niskie} zł")
print(f"Procent całego magazynu: {procent_magazynu:.2f}%")
```

## Zadanie 6 (bonus): Analiza korelacji cena–ocena
- Które produkty są "okazją" — ocena >= 4.5 i cena < 1000 zł?  
- Jaka jest średnia cena produktów z oceną >= 4.0 vs < 4.0?

```python
# Zadanie 6 (bonus): Analiza korelacji cena–ocena
mask_okazje = (oceny_klientow >= 4.5) & (ceny < 1000)
print("Produkty 'okazja' (ocena >= 4.5 i cena < 1000 zł):")
for n, c in zip(nazwy_array[mask_okazje], ceny[mask_okazje]):
    print(f"{n}: {c} zł")

# Średnia cena produktów ocena >=4.0 vs <4.0
mask_ge4 = oceny_klientow >= 4.0
mask_lt4 = oceny_klientow < 4.0
srednia_ge4 = ceny[mask_ge4].mean()
srednia_lt4 = ceny[mask_lt4].mean()

print(f"\nŚrednia cena produktów z oceną >= 4.0: {srednia_ge4:.2f} zł")
print(f"Średnia cena produktów z oceną < 4.0: {srednia_lt4:.2f} zł")
```

## Ćwiczenie 4:

```python
import numpy as np
import time

# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)

# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start

print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start

start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start

print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start

print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
