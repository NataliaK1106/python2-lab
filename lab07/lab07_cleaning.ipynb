## Ćwiczenie 3: Pełny pipeline czyszczenia

```python
df_clean = pd.DataFrame(data)  # świeży start

# KROK 1: wynagrodzenie → NaN + float
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].replace('brak', np.nan)
df_clean['wynagrodzenie'] = pd.to_numeric(df_clean['wynagrodzenie'], errors='coerce')

# KROK 2: usunięcie duplikatów + reset indeksu
df_clean = df_clean.drop_duplicates().reset_index(drop=True)

# KROK 3: wypełnienie wynagrodzenia medianą
mediana = df_clean['wynagrodzenie'].median()
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].fillna(mediana)

# KROK 4: ocena_roczna → średnia
srednia = round(df_clean['ocena_roczna'].mean(), 2)
df_clean['ocena_roczna'] = df_clean['ocena_roczna'].fillna(srednia)

# KROK 5: czyszczenie imion
df_clean['imie'] = df_clean['imie'].str.strip().str.title()

# KROK 6: czyszczenie działów
df_clean['dzial'] = df_clean['dzial'].str.strip().str.title()
df_clean['dzial'] = df_clean['dzial'].str.replace('Hr', 'HR', regex=False)
df_clean['dzial'] = df_clean['dzial'].str.replace('It', 'IT', regex=False)

# KROK 7: konwersja daty
df_clean['data_zatrudnienia'] = pd.to_datetime(
    df_clean['data_zatrudnienia'],
    errors='coerce'
)

# ======================
# WERYFIKACJA
# ======================
print("=== WERYFIKACJA ===")
print(f"Shape: {df_clean.shape}")
print(f"\nDtypes:\n{df_clean.dtypes}")
print(f"\nNaN per kolumna:\n{df_clean.isna().sum()}")
print(f"\nUnikalne działy: {sorted(df_clean['dzial'].unique())}")
print(f"\nPierwsze imiona: {df_clean['imie'].tolist()[:8]}")
```

```python
# Pytanie 1: Jakie jest średnie wynagrodzenie per dział?
df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2)
```

```python
# Pytanie 2: Jaka jest średnia ocena per dział?
df_clean.groupby('dzial')['ocena_roczna'].mean().round(2)
```

```python
# Pytanie 3: Kto jest najlepiej opłacany (top 3)?
df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial', 'wynagrodzenie']]
```

```python
# Pytanie 4: Ilu pracowników jest w każdym dziale?
df_clean['dzial'].value_counts()
```

```python
# Pytanie 1: Ile lat pracuje najdłuższy staż? (dzisiejsza data - najwcześniejsza data zatrudnienia)

najwczesniejsza_data = df_clean['data_zatrudnienia'].min()
staz_lat = (pd.Timestamp.today() - najwczesniejsza_data).days / 365

print(round(staz_lat, 1))
```

```python
# Pytanie 2: Kto ma najwyższą ocenę w każdym dziale? (groupby + idxmax)

top_oceny = df_clean.loc[
    df_clean.groupby('dzial')['ocena_roczna'].idxmax(),
    ['imie', 'dzial', 'ocena_roczna']
]

print(top_oceny)
```

```python
# Pytanie 3: Czy jest korelacja między stażem a wynagrodzeniem?

df_clean['staz_lat'] = (pd.Timestamp.today() - df_clean['data_zatrudnienia']).dt.days / 365

korelacja = df_clean[['staz_lat', 'wynagrodzenie']].corr().iloc[0,1]

print(round(korelacja, 3))
```
